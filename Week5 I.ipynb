```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
data = pd.read_csv('Banknote-authentication-dataset-labelled.csv')
mean = np.mean(data)
std = np.std(data)
v1 = data['V1']
v2 = data['V2']
data.head()
```

```python
plt.xlabel("Variance")
plt.ylabel('Skewness')
v1_v2 = np.column_stack((v1,v2))
km_res = KMeans(n_clusters=2).fit(data)
clusters = km_res.cluster_centers_

groups = data.groupby("Class")
for name, group in groups:
    plt.scatter(group["V1"], group["V2"], marker="o", label = name,alpha=0.50)
plt.legend(('Genuine','Forged'))

plt.scatter(clusters[:, 0],clusters[:, 1], s=250, c = 'black', marker='p')
```
